3) Create a new column/variable named as "LargeClaim_Flag" = "Yes" when TotalIncurredAsOf2YearsFromClaimDateOpen >=40000;  "LargeClaim_Flag" = "No" when TotalIncurredAsOf2YearsFromClaimDateOpen <40000 (10 Points) 

4) Group by the new variable/column ClaimStstus with mean of the TotalIncurredAsOf2YearsFromClaimDateOpen column (5 Points) 

5) Create a frequency table of the new variable/column of LargeClaim_Flag (Hint: use .value_counts()) (5 Points)

```python
Cdata.head(2)
#Cdata.shape
```

```python
# 1) Type your answers here
WCClosed = Cdata[(Cdata['LineCode'] == 'WC') & (Cdata['ClaimStatus'] == 'C')]
WCClosed.head()
WCClosed.shape
```

```python
# 2) Type your answers here
WCClosedAcc = WCClosed[WCClosed['ClaimSubstatus'] == 'Accepted']
WCClosedAcc.head()
WCClosedAcc.shape
```

```python
# 3) Type your answers here
# Assume this new column is established based on the Cdata instead of WCClosedAcc data
# If this data is extracted from the WCClosedAcc data, then just change Cdata to WCClosedAcc and get the answer. 

conditions = [(Cdata['TotalIncurredAsOf2YearsFromClaimDateOpen'] >= 40000),
              (Cdata['TotalIncurredAsOf2YearsFromClaimDateOpen'] <  40000)]

values = ['Yes', 'No']

Cdata['LargeClaim_Flag'] = np.select(conditions, values)

Cdata.head(2)
```

```python
# 4) Type your answers here
# Assume this new column is established based on the Cdata instead of WCClosedAcc data
# If this data is extracted from the WCClosedAcc data, then just change Cdata to WCClosedAcc and get the answer. 

group_C = Cdata.groupby("LargeClaim_Flag")
group_C[['TotalIncurredAsOf2YearsFromClaimDateOpen']].mean()
```

```python
# 5) Type your answers here
# Assume this new column is established based on the Cdata instead of WCClosedAcc data
# If this data is extracted from the WCClosedAcc data, then just change Cdata to WCClosedAcc and get the answer. 

Cdata['LargeClaim_Flag'].value_counts()
```

```python
# 6) Type your answers here
# Assume this new column is established based on the Cdata instead of WCClosedAcc data
# If this data is extracted from the WCClosedAcc data, then just change Cdata to WCClosedAcc and get the answer. 

WCFinal = Cdata[['PolicyNumber', 'ClaimType', 'StateofJurisdiction', 'LargeClaim_Flag', 
                 'TotalIncurredAsOf2YearsFromClaimDateOpen']]
WCFinal.head()
```

```python
# 7) Type your answers here
WCFinal.to_csv('WCFinal.csv',index=False)
```

### Question 13. Create a correlation table and then use it to plot a clustermap. (5 points)

```python
# Type Your Answer Here:
import pandas as pd
ct = pd.DataFrame(titanic,columns=['sibsp','parch','survived','fare','pclass','age','adult_male','alone'])
corrMatrix = ct.corr()
sbn.clustermap(corrMatrix, cmap='coolwarm')
```
